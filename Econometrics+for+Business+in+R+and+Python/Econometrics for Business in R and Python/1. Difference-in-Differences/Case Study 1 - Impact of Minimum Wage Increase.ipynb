Importing the libraries

```python
import numpy as np
import pandas as pd
```

pulling the data

```python
dataset = pd.read_csv("njmin3.csv")
```

checking the data

```python
dataset.describe()
pd.set_option("display.max_columns", None)
dataset
```

check if there are more NAs

```python
dataset.isnull().any()
```

replacing NAs with averages

```python
from sklearn.impute import SimpleImputer
missingvalues = SimpleImputer(missing_values = np.nan,
                              strategy = "mean")
missingvalues = missingvalues.fit(dataset[['fte', 'demp']])
dataset[['fte', 'demp']] = missingvalues.transform(dataset[['fte', 'demp']])
```

check if there are more NAs

```python
dataset.isnull().any()
```

```python
dataset
```

isolating the X and Y variables

```python
X = dataset.iloc[:,0:3].values
```

```python
X
```

```python
Y = dataset.iloc[:,3].values
```

```python
Y
```

creating the first model

```python
import statsmodels.api as sm
X = sm.add_constant(X)
model1 = sm.OLS(Y, X).fit()
```

```python
model1.summary(yname = "fte", 
               xname = ("intercept", "New Jersey", "After April 1992", "New Jersey and after April 1992"))
```

Isolating the X and Y variables part 2

```python
X = dataset.loc[:,['NJ', "POST_APRIL92", "NJ_POST_APRIL92", "bk", "kfc", "wendys"]].values
```

creating the second model

```python
X = sm.add_constant(X)
model2 = sm.OLS(Y, X).fit()
model2.summary(yname = "fte", 
               xname = ("intercept", "New Jersey", "After April 1992", "New Jersey and after April 1992", "Burger King", "KFC", "Wendy's"))
```

```python
X = dataset.loc[:,['NJ', "POST_APRIL92", "NJ_POST_APRIL92", "bk", "kfc", "wendys", "co_owned", "centralj", "southj" ]].values
```
